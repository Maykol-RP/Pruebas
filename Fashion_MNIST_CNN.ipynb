```python
# ==================================================
# CRISP-ML: Fashion-MNIST con CNN
# Importando directamente los archivos CSV
# ==================================================

# ==================================================
# FASE 1: Comprensión del negocio y de los datos
# ==================================================
# Objetivo: Clasificar imágenes de ropa en 10 categorías usando CNN

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.utils import to_categorical

# Cargar datos de Fashion-MNIST directamente desde Keras datasets
print("Cargando datos de Fashion-MNIST desde Keras...")
(X_train_raw, y_train_raw), (X_test_raw, y_test_raw) = keras.datasets.fashion_mnist.load_data()

# Convertir a DataFrame para mantener la estructura original del código (opcional, se puede refactorizar)
# Para simular la estructura de los CSV:
# train_df tendrá 'label' y luego columnas de píxeles
# test_df tendrá 'label' y luego columnas de píxeles

# Aplanar las imágenes y combinarlas con las etiquetas
train_pixels = X_train_raw.reshape(X_train_raw.shape[0], -1)
test_pixels = X_test_raw.reshape(X_test_raw.shape[0], -1)

train_df = pd.DataFrame(train_pixels)
train_df.insert(0, 'label', y_train_raw)

test_df = pd.DataFrame(test_pixels)
test_df.insert(0, 'label', y_test_raw)

# Los CSV tienen: primera columna = label, luego 784 columnas de pixeles (28x28)
# print("Cargando datos desde CSV...")
# train_df = pd.read_csv('fashion-mnist_train.csv')  # 60,000 filas
# test_df = pd.read_csv('fashion-mnist_test.csv')    # 10,000 filas

# Exploración inicial
print(f"\nTrain set shape: {train_df.shape}")
print(f"Test set shape: {test_df.shape}")
print(f"\nPrimeras 5 filas del train set:")
print(train_df.head())
print(f"\nInformación del dataset:")
print(train_df.info())

print(f"\nColumnas: {train_df.columns[:5]} ... {train_df.columns[-5:]}")
print(f"Rango de valores de píxeles: min={train_df.iloc[:,1:].min().min()}, max={train_df.iloc[:,1:].max().max()}")

# Nombres de las clases
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# Ver distribución de clases
print("\nDistribución de clases en entrenamiento:")
label_counts = train_df['label'].value_counts().sort_index()
for i, count in label_counts.items():
    print(f"  {i} - {class_names[i]}: {count} imágenes")

# Visualizar algunas imágenes
def show_images_from_df(df, num_images=25):
    plt.figure(figsize=(12,12))
    for i in range(num_images):
        plt.subplot(5,5,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)

        # Obtener imagen (primeras 784 columnas después de label)
        img = df.iloc[i, 1:].values.reshape(28, 28)
        label = df.iloc[i, 0]

        plt.imshow(img, cmap='gray')
        plt.xlabel(class_names[label])
    plt.suptitle("Ejemplos de Fashion-MNIST")
    plt.show()

show_images_from_df(train_df)

# ==================================================
# FASE 2: Preparación de datos
# ==================================================
def prepare_data(df, is_train=True):
    """
    Prepara los datos desde DataFrame de CSV
    Retorna: (imágenes normalizadas, labels one-hot encoded)
    """
    # Separar labels y pixeles
    labels = df['label'].values
    images = df.drop('label', axis=1).values

    # Normalizar pixeles (0-255 -> 0-1)
    images = images.astype('float32') / 255.0

    # Reshape a formato de imagen 28x28x1 (canal gris)
    images = images.reshape(-1, 28, 28, 1)

    # Convertir labels a one-hot encoding
    labels = to_categorical(labels, 10)

    return images, labels

# Preparar datos de entrenamiento y prueba
print("\nPreparando datos...")
X_train, y_train = prepare_data(train_df)
X_test, y_test = prepare_data(test_df)

# Dividir entrenamiento en train/validation (80/20)
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
)

print(f"\nForma de los datos después de preparación:")
print(f"  Train: {X_train.shape}, Labels: {y_train.shape}")
print(f"  Validation: {X_val.shape}, Labels: {y_val.shape}")
print(f"  Test: {X_test.shape}, Labels: {y_test.shape}")

# Verificar rango de valores
print(f"Rango de valores de píxeles después de normalización: [{X_train.min()}, {X_train.max()}]")

# ==================================================
# FASE 3: Modelado (CNN)
# ==================================================
def create_cnn_model():
    """
    Arquitectura CNN optimizada para Fashion-MNIST
    """
    model = models.Sequential([
        # Primera capa convolucional
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),

        # Segunda capa convolucional
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),

        # Tercera capa convolucional
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.BatchNormalization(),

        # Capas densas
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(10, activation='softmax')
    ])

    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# Crear y mostrar modelo
model = create_cnn_model()
print("\nArquitectura del modelo:")
model.summary()

# Callbacks para mejor entrenamiento
callbacks = [
    keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True,
                                  monitor='val_accuracy'),
    keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=3,
                                      monitor='val_loss'),
    keras.callbacks.ModelCheckpoint('best_fashion_mnist_model.h5',
                                    save_best_only=True,
                                    monitor='val_accuracy')
]

# Entrenar el modelo
print("\nIniciando entrenamiento...")
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=30,
    batch_size=64,
    callbacks=callbacks,
    verbose=1
)

# ==================================================
# FASE 4: Evaluación
# ==================================================
# Evaluar en conjunto de prueba
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"\n=== Resultados en conjunto de prueba ===")
print(f"Pérdida (Loss): {test_loss:.4f}")
print(f"Precisión (Accuracy): {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")

# Predicciones
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

# Matriz de confusión
cm = confusion_matrix(y_true, y_pred_classes)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.title('Matriz de Confusión - Fashion-MNIST')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()

# Reporte de clasificación detallado
print("\n=== Reporte de Clasificación ===")
print(classification_report(y_true, y_pred_classes, target_names=class_names))

# Curvas de aprendizaje
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Entrenamiento', linewidth=2)
plt.plot(history.history['val_accuracy'], label='Validación', linewidth=2)
plt.title('Precisión del Modelo', fontsize=14)
plt.xlabel('Épocas')
plt.ylabel('Precisión')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Entrenamiento', linewidth=2)
plt.plot(history.history['val_loss'], label='Validación', linewidth=2)
plt.title('Pérdida del Modelo', fontsize=14)
plt.xlabel('Épocas')
plt.ylabel('Pérdida')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ==================================================
# FASE 5: Despliegue (simulado) y análisis adicional
# ==================================================
# Guardar modelo final
model.save('fashion_mnist_final_model.h5')
print("\nModelo guardado como 'fashion_mnist_final_model.h5'")

# Función para predecir imágenes específicas
def predict_single_image(image_array, model, class_names):
    """
    Predice una sola imagen (28x28 array)
    """
    if image_array.max() > 1:
        image_array = image_array / 255.0
    image_input = image_array.reshape(1, 28, 28, 1)
    prediction = model.predict(image_input, verbose=0)
    pred_class = np.argmax(prediction)
    confidence = np.max(prediction)
    return pred_class, confidence, prediction[0]

# Mostrar algunas predicciones del conjunto de prueba
num_examples = 10
plt.figure(figsize=(15,6))
for i in range(num_examples):
    plt.subplot(2,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)

    img = X_test[i].reshape(28,28)
    plt.imshow(img, cmap='gray')

    pred_class, confidence, probs = predict_single_image(img, model, class_names)
    true_class = y_true[i]

    color = 'green' if pred_class == true_class else 'red'
    plt.title(f'Pred: {class_names[pred_class]}\nReal: {class_names[true_class]}\nConf: {confidence:.2f}',
              color=color, fontsize=9)
plt.suptitle('Predicciones del Modelo (Verde = Correcto, Rojo = Incorrecto)', fontsize=14)
plt.tight_layout()
plt.show()

# Análisis de errores comunes
errors = np.where(y_pred_classes != y_true)[0]
print(f"\n=== Análisis de Errores ===")
print(f"Total de errores: {len(errors)} de {len(y_true)} ({len(errors)/len(y_true)*100:.2f}%)")

# Mostrar confusiones más comunes
error_pairs = [(y_true[i], y_pred_classes[i]) for i in errors[:20]]
unique_errors, counts = np.unique(error_pairs, return_counts=True, axis=0)
print("\nErrores más frecuentes:")
for (true, pred), count in zip(unique_errors[:5], counts[:5]):
    print(f"  {class_names[true]} → {class_names[pred]}: {count} veces")

# Ejemplo de imágenes mal clasificadas
plt.figure(figsize=(12,8))
for i, error_idx in enumerate(errors[:12]):
    plt.subplot(3,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)

    img = X_test[error_idx].reshape(28,28)
    plt.imshow(img, cmap='gray')
    plt.title(f'Real: {class_names[y_true[error_idx]]}\nPred: {class_names[y_pred_classes[error_idx]]}',
              color='red', fontsize=9)
plt.suptitle('Ejemplos de Imágenes Mal Clasificadas', fontsize=14)
plt.tight_layout()
plt.show()

print("\n=== CRISP-ML Completado Exitosamente ===")
print("Fases ejecutadas:")
print("1. Comprensión del negocio/datos - OK")
print("2. Preparación de datos - OK")
print("3. Modelado (CNN) - OK")
print("4. Evaluación - OK")
print("5. Despliegue (simulado) - OK")
```

```python
from google.colab import files
from PIL import Image
import io

print("Por favor, sube una imagen de tu computadora:")
uploaded_files = files.upload()

# Solo esperamos una imagen, tomamos la primera
if uploaded_files:
    uploaded_filename = next(iter(uploaded_files))
    print(f"Imagen subida: {uploaded_filename}")
    uploaded_image_pil = Image.open(io.BytesIO(uploaded_files[uploaded_filename]))

    plt.figure(figsize=(2,2))
    plt.imshow(uploaded_image_pil, cmap='gray')
    plt.title("Imagen Subida")
    plt.axis('off')
    plt.show()
else:
    print("No se subió ninguna imagen.")
    uploaded_image_pil = None # Asegúrate de que la variable no esté definida si no hay subida
```

### Preprocesar la Imagen Subida (Método Alternativo)

```python
if uploaded_image_pil is not None:
    # Convertir la imagen PIL a formato compatible con el modelo
    processed_image = uploaded_image_pil.convert('L') # Convertir a escala de grises
    processed_image = processed_image.resize((28, 28), Image.Resampling.LANCZOS)
    processed_image_array = np.array(processed_image)

    # Asegurarse de que el rango de píxeles sea 0-255 antes de normalizar si la imagen no se carga directamente así
    if processed_image_array.max() > 1:
        processed_image_array = processed_image_array.astype('float32') / 255.0

    processed_image_array = processed_image_array.reshape(1, 28, 28, 1)

    print(f"Forma de la imagen preprocesada: {processed_image_array.shape}")
    print(f"Rango de valores de píxeles: [{processed_image_array.min()}, {processed_image_array.max()}]")

    plt.figure(figsize=(2,2))
    plt.imshow(processed_image_array.reshape(28,28), cmap='gray')
    plt.title("Imagen Preprocesada")
    plt.axis('off')
    plt.show()
else:
    print("No hay imagen subida para preprocesar. Por favor, sube una imagen en la celda anterior.")
    processed_image_array = None # Asegúrate de que la variable no esté definida
```

### Realizar Predicción con la Imagen Subida (Método Alternativo)

```python
if 'processed_image_array' in locals() and processed_image_array is not None:
    pred_class_idx, confidence, all_probabilities = predict_single_image(
        processed_image_array.reshape(28,28), model, class_names
    )

    predicted_class_name = class_names[pred_class_idx]

    print(f"\nLa predicción del modelo para la imagen subida es: '{predicted_class_name}'")
    print(f"Con una confianza de: {confidence:.2f}")

    plt.figure(figsize=(3,3))
    plt.imshow(processed_image_array.reshape(28,28), cmap='gray')
    plt.title(f"Predicción: {predicted_class_name}\nConfianza: {confidence:.2f}", fontsize=12, color='blue')
    plt.axis('off')
    plt.show()

    # Opcional: Mostrar todas las probabilidades
    print("\nProbabilidades para cada clase:")
    for i, prob in enumerate(all_probabilities):
        print(f"  {class_names[i]}: {prob:.4f}")
else:
    print("Error: La imagen no ha sido subida o preprocesada correctamente. Por favor, asegúrate de ejecutar las celdas anteriores de 'Subir imagen' y 'Preprocesar imagen'.")
```
